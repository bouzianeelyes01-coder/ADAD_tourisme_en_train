```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# ----------------------------------------
# Lecture des CSV
# ----------------------------------------
tgvmax = pd.read_csv(
    "tgvmax.csv",
    sep=';',
    dtype=str
)

qualite = pd.read_csv(
    "etablissements-labellises-qualite-tourisme.csv",
    sep=';',
    dtype=str
)

```

```python
# ----------------------------------------
# Filtrer les destinations depuis Marseille
# ----------------------------------------
tgv_marseille = tgvmax[
    (tgvmax["Origine"] == "MARSEILLE ST CHARLES") &
    (tgvmax["Disponibilité de places MAX JEUNE et MAX SENIOR"] == "OUI")
].copy()

# Garder destination + durée
tgv_marseille["Destination_clean"] = tgv_marseille["Destination"].str.upper().str.strip()

```

```python
# ----------------------------------------
# Nettoyage des noms de destinations
# ----------------------------------------
def clean_city(x):
    if pd.isna(x):
        return None
    x = x.upper()
    x = re.sub(r"\s*(TGV|CENTRE|VILLE|MIDI|MATABIAU|CHANTIERS|INTRAMUROS)\s*", "", x)
    x = re.sub(r"\bSAINT\b", "ST", x)
    x = re.sub(r"\bSAINTE\b", "STE", x)
    x = re.sub(r"[()]", "", x)
    return x.strip()

tgv_marseille["Destination_clean"] = tgv_marseille["Destination_clean"].apply(clean_city)

special = {
    "ST RAPHAEL": "SAINT-RAPHAEL",
    "AIX EN PROVENCE": "AIX-EN-PROVENCE",
    "CHALON SUR SAONE": "CHALON-SUR-SAONE"}

tgv_marseille["Destination_clean"] = tgv_marseille["Destination_clean"].replace(special)

```
